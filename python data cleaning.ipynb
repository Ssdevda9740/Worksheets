```python
# 9. Determine the column names having more than 30% missing values and remove them from the dataset.
```

```python
df_n=df.isnull().sum()
col=df.columns
for i in range(len(df_n)):
    if df_n[i]>7110:  #7110 is 30% of total data
        print(i,col[i],':'," REMOVE the dataset \n")
    else:
        print(i,col[i],':', 'Need to use mean or mode to fill \n')
```

```python
df_n=df.isnull().sum()
col=df.columns
for i in range(len(df_n)):
    if df_n[i]>7110:  #7110 is 30% of total data
        print(i,col[i],':'," REMOVE the dataset \n")
        df[col[i]]=df[col[i]].dropna()
    else:
        print(i,col[i],':', 'Need to use mean or mode to fill \n')
```

```python
df['BuildingArea']=df['BuildingArea'].dropna()
df['YearBuilt']=df['YearBuilt'].dropna()
df['CouncilArea']=df['CouncilArea'].dropna()
```
